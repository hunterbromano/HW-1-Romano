```python
# It is intertesting to note that the more fans that are in attendance, the less goals will be scored
# I want to look at how attendance changed over the years
# Some data points were empty and so I had to drop those values with missing = 'drop'

reg = sm.OLS(endog=fifa['Year'], exog = fifa['Attendance'], missing = 'drop').fit()
reg.summary()

# r squared is high so the correlation is high.
```

```python
# Create a scatter plot to look at the year vs attendance relationship discussed earlier.

fig, ax = plt.subplots()

ax.scatter(x = fifa['Year'], y = fifa['Attendance'])

# Add Labels and a Title, while customizing font

plt.ylabel('Attendance', fontsize = 14)
plt.xlabel('Year', fontsize = 14)
plt.title ('World Cup Attendance Over the Years', fontsize = 16)

# Clean up code above the plot

plt.show()

```

```python
# See what the country of origin means for the price of the wine on average

wine.groupby('country')['price'].mean()
```

```python
# I want to rank this list by price for easier visualization
# To do that I must make it a data frame

pd.DataFrame(wine.groupby('country')['price'].mean())
```

```python
# sort the new data frame by price

wine_price = pd.DataFrame(wine.groupby('country')['price'].mean())

wine_price
```

```python
#wine_price by highest to lowest price and dig down to the top 10

wine_top10 = wine_price.sort_values(by ='price',ascending = False).head(10)

wine_top10
```

```python
# Now I want to generate some plots to better visualize the data

wine_top10 = wine_top10.reset_index()
```

```python
sns.barplot(wine_top10.country, wine_top10.price)

# Add a title

plt.title("Avg. Wine Price by Country")

# unbunch the x axis variables

plt.xticks(rotation=90)
```
